```python
import pandas as pd


df = pd.read_csv('/content/Unemployment in India.csv')


df.head()
```

```python

df.info()
df.isnull().sum()


df.columns


df['Region'].unique()
```

```python
print(df.columns)
```

```python

df.rename(columns=lambda x: x.strip(), inplace=True)


df['Date'] = pd.to_datetime(df['Date'])
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))
sns.lineplot(x='Date', y='Estimated Unemployment Rate (%)', data=df)
plt.title('Unemployment Rate in India Over Time')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.show()
```

```python
plt.figure(figsize=(14,6))
sns.lineplot(x='Date', y='Estimated Unemployment Rate (%)', hue='Region', data=df)
plt.title('State-wise Unemployment Trend During COVID-19')
plt.xticks(rotation=45)
plt.show()
```

```python

covid_df = df[df['Date'].dt.year == 2020]


avg_2020 = covid_df.groupby('Region')['Estimated Unemployment Rate (%)'].mean().sort_values(ascending=False)

plt.figure(figsize=(10,5))
avg_2020.plot(kind='bar', color='crimson')
plt.title("Average Unemployment During COVID-19 (2020)")
plt.ylabel("Unemployment Rate (%)")
plt.xticks(rotation=90)
plt.show()
```

```python
import pandas as pd


df = pd.read_csv('/content/Unemployment_Rate_upto_11_2020.csv')


print(df.columns)


df.head()
```

```python

df.columns = df.columns.str.strip()


print(df.columns)
```

```python

df['Date'] = pd.to_datetime(df['Date'])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))
sns.lineplot(x='Date', y='Estimated Unemployment Rate (%)', data=df)
plt.title('Unemployment Rate in India (Till Nov 2020)')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(14,6))
sns.lineplot(x='Date', y='Estimated Unemployment Rate (%)', hue='Region', data=df)
plt.title('State-wise Unemployment Trends (Till Nov 2020)')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.legend(loc='upper right')
plt.show()
```

```python

df_2020 = df[df['Date'].dt.year == 2020]


avg_2020 = df_2020.groupby('Region')['Estimated Unemployment Rate (%)'].mean().sort_values(ascending=False)

plt.figure(figsize=(12,6))
avg_2020.plot(kind='bar', color='orange')
plt.title("Average Unemployment by Region (2020)")
plt.ylabel("Unemployment Rate (%)")
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```
